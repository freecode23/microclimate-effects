```python
import pandas as pd
import numpy as np
import pathlib
import pickle
from datetime import datetime
import os
```

Use ENVI-met (Raw) file from google drive

```python
bld_labels = pd.read_csv('../microclimate_model/Building Labels.csv')
```

```python
bld_labels['Building Name'] = bld_labels['Building Name'].astype(str)
bld_labels
```

```python
# ENVIMET DATA #
ENV = []
for path in pathlib.Path("../microclimate_model/ENVI-met").iterdir():
    current_folder_date = path.name
    for path in pathlib.Path("../microclimate_model/ENVI-met" + current_folder_date + "/base_micro").iterdir():
        if path.is_file():
            current_file = str(path)
            try:
                current_file_number = int(current_file[-8:-4])
            except ValueError:
                pass
        for i in range(len(bld_labels)):
            if current_file_number == bld_labels['Building Nbr'][i]:
                current_file_name = bld_labels['Building Name'][i]
        new_file = current_file[:99] + current_file_name + '.csv'
        os.rename(current_file, new_file)
```
