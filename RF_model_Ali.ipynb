```python
import pandas as pd
import pathlib
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.pyplot import cm
```

```python
############################## READ ALL DATA #############################################
## WEATHER FILES ##
# AZ PHX Sky Harbor Data #
AZW_15 = pd.read_csv("./Data/Weather Data/KPHX-15.csv")
# ENVIMET DATA #
BPS = []
Fname = []
for path in pathlib.Path("./Data/BPS").iterdir():
    if path.is_file():
        current_file = pd.read_csv(path)
        BPS.append(current_file)
        Fname.append(path.name.replace('.csv', ''))

## Drop last row of EnviMet Data
for i in range(len(BPS)):
    BPS[i] = BPS[i].drop(16)
```

```python
######################### CHOOSNG ENVI-MET BUILDING MICROCLIMATE DATA #################################

## BUILDING ##

# Interdisciplinary AB Microclimate Data
Inter_AB = pd.read_csv("./Data/Weather Data/Interdisciplinary AB 2018.csv")
Inter_AB_W = pd.DataFrame()
Inter_AB_W = Inter_AB_W.append(BPS[7])
Inter_AB_W = Inter_AB_W[['Time', 'AirTempInFrontOfAllFacades_MEAN', 'RelativeAirHumidityInFrontOfAllFacades_MEAN',
                     'WindSpeedInFrontOfAllFacades_MEAN']]

Inter_AB_W = Inter_AB_W.rename(columns = {'AirTempInFrontOfAllFacades_MEAN':'Mean Air Temp',
                                    'RelativeAirHumidityInFrontOfAllFacades_MEAN':'Mean Rel Humidity',
                                    'WindSpeedInFrontOfAllFacades_MEAN':'Mean Wind Speed'})
Inter_AB_W
```

```python
########################### FORMATING BIG DATA FRAMES ##############################

## Weather file ##
AZW_15 = AZW_15.set_index('Date_Time')


## Building ##
# Inter_AB = Interdisciplinary AB Campus Metabolism Data
# Data_Inter = Inter_AB + Weather Data

##### Step 1
for i in range(len(Inter_AB)):
    Inter_AB['tstamp'][i] = (str(Inter_AB['tstamp'][i][5:7]) + '/' + str(Inter_AB['tstamp'][i][8:10]) + '/2018' + ' ' + str(Inter_AB['tstamp'][i][11:16]))
    
Inter_AB = Inter_AB[['tstamp', 'bldgname', 'KW', 'CHWTON', 'HTmmBTU']]

##### Step 2 
month = []
time = []

for i in range(len(Inter_AB)):
    month.append(Inter_AB['tstamp'][i][0:2])

for i in range(len(Inter_AB)):
    time.append((Inter_AB['tstamp'][i][11:13] + Inter_AB['tstamp'][i][14:]))

Inter_AB['month'] = month
Inter_AB['time'] = time


##### Step 3
Inter_AB = Inter_AB.set_index('tstamp')


##### Step 4
Data_Inter =  pd.concat([Inter_AB, AZW_15], axis=1, join="inner")
Data_Inter = Data_Inter[['month', 'time', 'bldgname', 'KW', 'CHWTON', 'HTmmBTU', 'Air Temp', 'Rel Humid']]
```

```python
######################## DATA FORMATTING for JUNE 9 ##########################

time = []
time.append('0500')
time.append('0600')
time.append('0700')
time.append('0800')
time.append('0900')
time.append('1000')
time.append('1100')
time.append('1200')
time.append('1300')
time.append('1400')
time.append('1500')
time.append('1600')
time.append('1700')
time.append('1800')
time.append('1900')
time.append('2000')

month = []
for i in range(16):
    month.append('06')

date = []
date.append('06/09/2018 05:00')
date.append('06/09/2018 06:00')
date.append('06/09/2018 07:00')
date.append('06/09/2018 08:00')
date.append('06/09/2018 09:00')
date.append('06/09/2018 10:00')
date.append('06/09/2018 11:00')
date.append('06/09/2018 12:00')
date.append('06/09/2018 13:00')
date.append('06/09/2018 14:00')
date.append('06/09/2018 15:00')
date.append('06/09/2018 16:00')
date.append('06/09/2018 17:00')
date.append('06/09/2018 18:00')
date.append('06/09/2018 19:00')
date.append('06/09/2018 20:00')

# Interdisciplinary AB June 9 Microclimate Variables Data #

Int_AB_CHWTON = []
x = 15170
for i in range(16):
    Int_AB_CHWTON.append(Data_Inter['CHWTON'][i + x])
    x = x + 3

Int_AB_KW = []
x = 15170
for i in range(16):
    Int_AB_KW.append(Data_Inter['KW'][i + x])
    x = x + 3

Int_AB_June_9 = pd.DataFrame()
Int_AB_June_9['month'] = month
Int_AB_June_9['time'] = time
Int_AB_June_9['CHWTON'] = Int_AB_CHWTON
Int_AB_June_9['KW'] = Int_AB_KW
Int_AB_June_9['date'] = date
Int_AB_June_9['Air Temp'] = Inter_AB_W['Mean Air Temp']
Int_AB_June_9['Rel Humid'] = Inter_AB_W['Mean Rel Humidity']

Int_AB_June_9 = Int_AB_June_9.set_index(['date'])
#Int_AB_June_9


# Interdisciplinary AB June 9 AZW Variables Data #

AZW_June_9_Air_Temp = []

x = 15170
for i in range(16):
    AZW_June_9_Air_Temp.append(Data_Inter['Air Temp'][i + x])
    x = x + 3

AZW_June_9_Rel_Hum = []
    
x = 15170
for i in range(16):
    AZW_June_9_Rel_Hum.append(Data_Inter['Rel Humid'][i + x])
    x = x + 3

Int_AB_June_9_AZW = pd.DataFrame()
Int_AB_June_9_AZW['month'] = month
Int_AB_June_9_AZW['time'] = time
Int_AB_June_9_AZW['CHWTON'] = Int_AB_CHWTON
Int_AB_June_9_AZW['KW'] = Int_AB_KW
Int_AB_June_9_AZW['date'] = date
Int_AB_June_9_AZW['Air Temp'] = AZW_June_9_Air_Temp
Int_AB_June_9_AZW['Rel Humid'] = AZW_June_9_Rel_Hum

Int_AB_June_9_AZW = Int_AB_June_9_AZW.set_index(['date'])
```

```python
############################ RANDOM FORESTS #################################

## Set Dependent Variable
Y = Data_Inter['CHWTON']

## Set Independent Variables
X = Data_Inter.drop(labels = ['CHWTON', 'HTmmBTU', 'bldgname'], axis = 1)

## Setting up training and testing data
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=20)

## Using RandomForestRegressor to make predictions
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X_train, Y_train)

print("\033[1m" + "Model Accuracy = " + "\033[0m", model.score(X_test, Y_test),"\n")

## Get feature importance
feature_list = list(X.columns)
feature_imp = pd.Series(model.feature_importances_, index=feature_list).sort_values(ascending=False)
print("\033[1m" + "Feature Importances:" + "\033[0m")
print(feature_imp)

## Predictions
Y_pred = model.predict(X_test)
ModelPred = pd.DataFrame({'Actual CHWTON':Y_test, 'Predicted CHWTON':Y_pred})
ModelPred = ModelPred.sort_index()
print(ModelPred)
```
