```python
############################## CHOOSE BUILDING #############################################

## Print Building Names ##
for i in range(len(Fname)):
    print(Fname[i])
    
bldname = input('Enter building name: ')

for i in range(len(BPS)):
    if bldname == Fname[i]:
        BPS_bldg = BPS[i]
```

```python
################################# CLASS TO FORMAT BUILDING DATA #############################################

class building:
    
    def __init__(self, bldgname, cmp, env):
        self.bldgname = bldgname
        self.cmp = cmp
        self.env = env
    
    def campusmetabolism(self):
        self.cmp = pd.read_csv("C:/Users/alyak/Desktop/ASU PhD/RA/Python/Data/ASU 2018/" + bldname + " 2018.csv")
        return self.cmp

    def envimet(self):
        self.env = BPS_bldg[['Time', 'AirTempInFrontOfAllFacades_MEAN', 'RelativeAirHumidityInFrontOfAllFacades_MEAN',
                     'WindSpeedInFrontOfAllFacades_MEAN']]
        self.env = self.env.rename(columns = {'AirTempInFrontOfAllFacades_MEAN':'Mean Air Temp',
                                    'RelativeAirHumidityInFrontOfAllFacades_MEAN':'Mean Rel Humidity',
                                    'WindSpeedInFrontOfAllFacades_MEAN':'Mean Wind Speed'})
        return self.env

    
cmp = pd.DataFrame()
env = pd.DataFrame()
Bldg = building(bldname, cmp, env)
cmp_data = Bldg.campusmetabolism()
env_data = Bldg.envimet()
```

```python
############################## FORMAT DATES #############################################

from astropy.time import Time

for i in range(len(cmp_data)):
    t = cmp_data['tstamp'][i][0:19]
    time1 = Time(t)
    cmp_data['tstamp'][i] = time1
```

```python
########################### FORMATING BIG DATA FRAMES ##############################

## Weather file ##
AZW_15 = AZW_15.set_index('Date_Time')


## Building ##
# cmp_data = Campus Metabolism Data
# all_data = cmp_data + AZ Weather Data

##### Step 1
for i in range(len(cmp_data)):
    cmp_data['tstamp'][i] = (str(cmp_data['tstamp'][i][5:7]) + '/' + str(cmp_data['tstamp'][i][8:10]) + '/2018' + ' ' + str(cmp_data['tstamp'][i][11:16]))
    
cmp_data = cmp_data[['tstamp', 'bldgname', 'KW', 'CHWTON', 'HTmmBTU']]

##### Step 2 
month = []
time = []

for i in range(len(cmp_data)):
    month.append(cmp_data['tstamp'][i][0:2])

for i in range(len(cmp_data)):
    time.append((cmp_data['tstamp'][i][11:13] + cmp_data['tstamp'][i][14:]))

cmp_data['month'] = month
cmp_data['time'] = time


##### Step 3
cmp_data = cmp_data.set_index('tstamp')


##### Step 4
all_data =  pd.concat([cmp_data, AZW_15], axis=1, join="inner")
all_data = all_data[['month', 'time', 'bldgname', 'KW', 'CHWTON', 'HTmmBTU', 'Air Temp', 'Rel Humid']]

##### DROP June 9th Weather
# row 15150 to row 15246 are all june 19th data
all_data = all_data.drop(all_data.index[15150:15246])
```
